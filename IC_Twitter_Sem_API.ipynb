# #paraisopolis

```python
# Store the results to a file in the JSON ******************************************************************************
import twint

c = twint.Config()
c.Search = '#paraisopolis'

c.Since = '2019-12-01'
c.Until = '2020-12-31'
c.Store_json = True
c.Output = "paraisopolis_2019.json"

twint.run.Search(c)
#***********************************************************************************************************************
```

```python
# Info *****************************************************************************************************************
df = pd.read_json('paraisopolis_2019.json', lines = True)
df.info()
#***********************************************************************************************************************
```

```python
# Plotting the frequency by day ****************************************************************************************
%matplotlib inline
df['date'].value_counts().plot()
#***********************************************************************************************************************
```

```python
# Plotting the frequency including retweets ****************************************************************************
df['tweet_counts'] = df['retweets_count'] + 1
df.groupby(df["date"]).sum()["tweet_counts"].plot()
#***********************************************************************************************************************
```

```python
# HTML to visualize the tweets *****************************************************************************************
from IPython.display import HTML
import requests

def show_tweet(link):
    '''Display the contents of a tweet. '''
    url = 'https://publish.twitter.com/oembed?url=%s' % link
    response = requests.get(url)
    html = response.json()["html"]
    display(HTML(html))
#***********************************************************************************************************************
```

```python
# Function HTML test ***************************************************************************************************
sample_tweet_link = df.sample(1)['link'].values[0]
display(sample_tweet_link)
show_tweet(sample_tweet_link)
#***********************************************************************************************************************
```

```python
# A list of the tweet urls, sorted by retweet count. *******************************************************************
rt_links = df.sort_values(by= 'retweets_count', ascending = False)['link'].values

for url in rt_links[:5]:
    print('üî• ' * 19)
    show_tweet(url)
#***********************************************************************************************************************
```
